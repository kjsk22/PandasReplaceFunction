```python
import pandas as pd
import numpy as np

df = pd.read_csv('EmployeeData.csv')

df
```

```python
df1 = df.replace(np.nan,0)
df1
```

```python
df1 = df1.replace(13,14)
df1
```

```python
df1 = df1.replace({'Age': 0}, 25)
df1
```

```python
df1 = df1.replace({'Age': 40.31, 'Experience': 0}, 30)
df1
```

```python
df1 = df1.replace({'Salary':0}, df1.Salary.mean())
df1
```

```python
df1 = df1.replace('Cigna','Aetna')
df1
```
